```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Get The data 
data =  pd.read_csv('D:\Python\Projects\Heart Disease Prediction\dataset\heart_disease.csv')
data
```

```python
data.drop(columns=['education'], inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['Gender'] = data['Gender'].replace(['Male', 'Female'], [1,0])
```

```python
data['prevalentStroke'] = data['prevalentStroke'].replace(['yes', 'no'], [1,0])
```

```python
data['Heart_ stroke'] = data['Heart_ stroke'].replace(['yes', 'No'], [1,0])
```

```python
data
```

```python
X = data.drop(columns=['Heart_ stroke'])
y = data['Heart_ stroke']
```

```python
y
```

### Imbalanced Dataset

```python
y.value_counts()
```

### To balance dataset

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
X, y = sm.fit_resample(X, y.ravel())
```

```python
np.unique(y, return_counts=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
input = np.array([[0, 59, 0,0,0,0,1,0,209,150,85,20.77,90,88]])
x = model.predict(input)
x[0]
```
